# Feature selection
##### 1-Variance threshold

```python
#Virer toutes les variables qui ont une faible variance : dont la valeur est la même sur 80% du train set threshold=(.8 * (1 - .8))
from sklearn.feature_selection import VarianceThreshold
```

```python
sel = VarianceThreshold(threshold=(.8 * (1 - .8)))
X_train_new = sel.fit_transform(X_train_with_dummies)
```

```python
X_test_new=sel.transform(X_test_with_dummies)
```

```python
print("Dimension du train set avant de virer les variables à faible variance   : ",X_train_with_dummies.shape)
print("Dimension du train set après avoir virer les variables à faible variance: ",X_train_new.shape)
```

```python
reg_after_feature_selection_variance = LinearRegression().fit(X_train_new, y_train_reshaped)
reg_after_feature_selection_variance.score(X_train_new, y_train_reshaped)
```

```python
print("Select subset of features using ""Variance threshold"" - Linear regression model - metrics over test set")
print ("R-squared : ",reg_after_feature_selection_variance.score(X_train_new, y_train_reshaped)) #coefficient de determination 1- (sum (yi-yi^)^2)/(sum(yi-moy(y))^2)
RSS = ((reg_after_feature_selection_variance.predict(X_train_new)-y_train_reshaped)**2).sum()
print("RSS : ", RSS)
print("MSE : ", RSS/(X_train_new.shape[0]))
```

```python
print("Select subset of features using ""Variance threshold"" - Linear regression model - metrics over test set")
print ("R-squared : ",reg_after_feature_selection_variance.score(X_test_new, y_test_reshaped)) #coefficient de determination 1- (sum (yi-yi^)^2)/(sum(yi-moy(y))^2)
RSS = ((reg_after_feature_selection_variance.predict(X_test_new)-y_test_reshaped)**2).sum()
print("RSS : ", RSS)
print("MSE : ", RSS/(X_test_new.shape[0]))
```

###### $conclusion$ : la selection de varibales en prenant en compte la faible variance donne un score de prediction plus faible sur les données d'entrainement mais plus élevé sur les données de test (comparée à la regression linéaire sur l'ensemble des variables du modèle)

```python
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
r_squared_train =[]
r_squared_test = []

rss_train =[]
rss_test = []

mse_train =[]
mse_test = []

for i in [15, 20, 25, 30, 35, 40, 45, 50, 55, 60] :
    reg_linear = LinearRegression()
    sfs1 = sfs(reg_linear,
               k_features=i,
               forward=True,
               floating=False,
               verbose=2,
               scoring='neg_mean_squared_error',
               cv=5)

    # Perform SFFS
    sfs1 = sfs1.fit(X_train_with_dummies, y_train_reshaped)
    
    # Which features?
    feat_cols = list(sfs1.k_feature_idx_)
    print(i," features : " ,feat_cols)
    
    reg_linear_1=LinearRegression()
    reg_linear_1.fit(X_train_with_dummies.iloc[:,feat_cols], y_train_reshaped)
    
    r_squared_train.append(reg_linear_1.score(X_train_with_dummies.iloc[:,feat_cols], y_train_reshaped))
    r_squared_test.append(reg_linear_1.score(X_test_with_dummies.iloc[:,feat_cols], y_test_reshaped))
    
    rss_train_tmp=((reg_linear_1.predict(X_train_with_dummies.iloc[:,feat_cols])-y_train_reshaped)**2).sum()
    rss_test_tmp = ((reg_linear_1.predict(X_test_with_dummies.iloc[:,feat_cols])-y_test_reshaped)**2).sum()
    rss_train.append(rss_train_tmp)
    rss_test.append(rss_test_tmp)
    
    mse_train.append(rss_train_tmp/(X_train_with_dummies.iloc[:,feat_cols].shape[0]))
    mse_test.append(rss_test_tmp/(X_test_with_dummies.iloc[:,feat_cols].shape[0]))
    
    
```

```python
#Residual sum is squares
def rss(X,y, model):
    rss = ((model.predict(X)-y)**2).sum()
    return rss
#mean squares error
def mse(X,y, model):
    return rss(X,y,model)/(X.shape[0])
```

```python
from sklearn.linear_model import Ridge
print(" ******** Ridge regression without feature selection, alpha = 0.5")
## train model using ridge regression using all variables
ridgeReg = Ridge(alpha=0.5, normalize=True) #Ridge(alpha=0.05, normalize=True)
ridgeReg.fit(X_train_with_dummies,y_train_reshaped)

print("R^2 train : ", ridgeReg.score(X_train_with_dummies,y_train_reshaped))
print("R^2 test : ", ridgeReg.score(X_test_with_dummies,y_test_reshaped))

print("RSS train : ", rss(X_train_with_dummies,y_train_reshaped, ridgeReg ))
print("RSS test : ", rss(X_test_with_dummies, y_test_reshaped, ridgeReg))

print("MSE train : ", mse(X_train_with_dummies,y_train_reshaped, ridgeReg ))
print("MSE test : ", mse(X_test_with_dummies, y_test_reshaped, ridgeReg))
```
